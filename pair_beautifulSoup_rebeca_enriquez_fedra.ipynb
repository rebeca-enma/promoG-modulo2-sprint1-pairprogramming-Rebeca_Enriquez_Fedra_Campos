```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
def scrape_coches_ocasion(url):

    respuesta_coches = requests.get(url)
    

    sopa_coches = BeautifulSoup(respuesta_coches.content,'html.parser')

    lista_nombre_coches = sopa_coches.find_all("h2", {"style": "overflow-x:hidden"})
    lista_precio_rebajado = sopa_coches.find_all("span", {"class": "jsx-2870186023 priceLast"})
    lista_precio_original = sopa_coches.find_all("span", {"class": "jsx-2689423372 price"})
    lista_info= sopa_coches.find_all("div", {"class": "jsx-2480038221 op-grid-card-info"})
    lista_km= []

    for coche in lista_info:
        busqueda= coche.find_all('span',{'class':'jsx-3960002075 detail big'})
        lista_km.append(busqueda[1].text)

    nombre_coches = [coche.getText() for coche in lista_nombre_coches]
    precio_original = [precio.getText() for precio in lista_precio_original]
    precio_rebajado = [precio.getText() for precio in lista_precio_rebajado ]

    encabezados = ["Coche", "Precio Original", "Precio Rebajado", "Kilometros"]

    df_coches = pd.DataFrame([nombre_coches,precio_original,precio_rebajado,lista_km]).T
    df_coches.columns = encabezados
    

    return df_coches

    
    

    
```

```python
ocasion_plus = scrape_coches_ocasion("https://www.ocasionplus.com/ofertas-especiales")
```
